## Task 1
### Meet Dr. Ignaz Semmelweis

In 1847, the Hungarian physician Ignaz Semmelweis makes a breakthough discovery: He discovers handwashing. Contaminated hands was a major cause of childbed fever and by enforcing handwashing at his hospital he saved hundreds of lives.

In this notebook, we will reanalyze the medical data Semmelweis collected.

```python
# Code for Task 1

# import modules
import pandas as pd

# Extracting data
yearly = pd.read_csv('datasets/yearly_deaths_by_clinic.csv')

# Output of Task 1
print(yearly)
```

## Task 2 
### The Alarming number of Deaths

```python
# Code for Task 2

# Calculate the proportion of deaths per number of births
yearly["proportion_deaths"] = yearly.deaths/yearly.births

# Extracting rows from clinic 1 and clinic 2
yearly1 = yearly[yearly.clinic == "clinic 1"]
yearly2 = yearly[yearly.clinic == "clinic 2"]

# Output of Task 2
print(yearly1)
```

## Task 3
### Death at the Clinics

```python
# Code for Task 3

# Import libraries to plot
import matplotlib.pyplot as plt
%matplotlib inline

# Plot the yearly proportion of deaths for both clinics
ax = yearly1.plot(x="year", y="proportion_deaths", label="Clinic 1")
yearly2.plot(x="year", y="proportion_deaths", label="Clinic 2", ax=ax)

# Labeling the axes
ax.set_xlabel("Year")
ax.set_ylabel("Proportion deaths")

# Title of plot
plt.title("Proportion of Deaths at Clinic 1 and Clinic 2")

# Output of Task 3
plt.legend()
plt.show()
```

## Task 4 
### The Handwashing begins

```python
# Code for Task 4

# import modules
import pandas as pd

# Extracting data
monthly = pd.read_csv('datasets/monthly_deaths.csv', parse_dates = ["date"])

# Calculate the proportion of deaths per number of births
monthly["proportion_deaths"] = monthly.deaths/monthly.births

# Output of task 4
monthly.head()
```

## Task 5 
### The Effect of Handwashing

```python
# Code for Task 5

# Import libraries to plot
import matplotlib.pyplot as plt
%matplotlib inline

# Plot the yearly proportion of deaths for both clinics
ax = monthly.plot(x='date', y="proportion_deaths", label="Proportion Deaths")

ax.set_xlabel("Date")
ax.set_ylabel("Proportion deaths")

plt.title(" Monthly proportion of Deaths ")

# Output for Task 5
plt.legend()
plt.show()
```

## Task 6
### The Effect of Handwashing highlighted

From above Monthly Proportions of Deaths plot, we can conclude that deaths have drastically reduced from summer of 1847 and Dr. Ignaz Semmelweis had successfully implemented handwashing.

```python
# Code for Task 6

# Observing above plot,Deaths have drastically reduced from summer of 1847 
handwashing_start = "1847-06-01"

# Split monthly into before and after handwashing_start
before_washing = monthly[monthly.date < handwashing_start]
after_washing = monthly[monthly.date >= handwashing_start]

# Import libraries to plot
import matplotlib.pyplot as plt
%matplotlib inline

# Plot the yearly proportion of deaths for both clinics
ax = before_washing.plot(x="date", y="proportion_deaths", label="Before Washing")
after_washing.plot(x="date", y="proportion_deaths", label="After Washing", ax=ax)

# Labeling the axes
ax.set_xlabel("Date")
ax.set_ylabel("Proportion deaths")

# Title of plot
plt.title("Highlights the Effect of Handwashing")

# Output of Task 6
plt.legend()
plt.show()
```

## Task 7
### More handwashing, fewer deaths?

```python
# Code for Task 7

#import library
import numpy as np

# Filtering required columns
before_proportion = before_washing.proportion_deaths
after_proportion = after_washing.proportion_deaths

# Difference in mean monthly proportion of deaths
mean_diff = np.mean(after_proportion) - np.mean(before_proportion)

#Output of Task 7
print(mean_diff)
```

## Task 8
### A bootstrap analysis of Semmelweis hand washing data

```python
# Code for Task 8

# A Bootstrap analysis of the difference in mean monthly proportion of deaths
boot_mean_diff = []
for i in range(3000):
    boot_before = before_proportion.sample(frac=1, replace=True)
    boot_after = after_proportion.sample(frac=1, replace=True)
    boot_mean_diff.append(np.mean(boot_after) - np.mean(boot_before))

# Calculate  confidence_interval as the quantiles of boot_mean_diff  
confidence_interval = pd.Series(boot_mean_diff).quantile([0.025, 0.975])

#Output of Task 8
print(np.array(confidence_interval))
```

## Task 9
### The fate of Dr. Semmelweis

So handwashing reduced the proportion of deaths by between 6.7 and 10 percentage points, according to a 95% confidence interval. All in all, it would seem that Semmelweis had solid evidence that handwashing was a simple but highly effective procedure that could save many lives.
